First, I will make a plot of the function.

```python
#Import necessary modules
import numpy as np
import matplotlib.pyplot as plt
import math

#Make x array
x = np.arange(0.2, 10, 0.01)

#Make y array from x
y = np.sin(x) - 0.4*np.cos(x)/x

#Plot graph and line showing x-axis
zero = x*0
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, zero)
ax.set(title='General Graph', xlabel='x', ylabel='y')
```

The first 3 roots seem to be around $x = 0.5$, $x = 3.5$, $x = 6.5$. The next section details finding a more exact value for the first root.

```python
#Make function for equation
def f(x):
    return math.sin(x) - 0.4*math.cos(x)/x

#Remake arrays to range around root
x = np.arange(0.2, 1, 0.01)
y = np.sin(x) - 0.4*np.cos(x)/x
zero = x*0

#Make plot, including axis and legend
fig, ax = plt.subplots()
ax.plot(x, y, label='function')
ax.plot(x, zero, label='x-axis')
ax.set(title='1st Root', xlabel='x', ylabel='y')
ax.legend(loc='best')

#Iterate using while loop and midpoint method to find root
#Plot iterations with dashed black lines.
xmin = 0.2
xmax = 1
root = (xmax - xmin)/2 + xmin
while f(root) > 1e-10 or f(root) < -1e-10:
    root = (xmax - xmin)/2 + xmin
    if f(root) < 0:
        xmin = root
    elif f(root) > 0:
        xmax = root
    ax.axvline(root, color='black', linestyle='dashed')

print('1st root: %s' %root)
```

```python
#Same as above method but with different range
#Adjusted for different slope when finding root
x = np.arange(3, 4, 0.01)
y = np.sin(x) - 0.4*np.cos(x)/x
zero = x*0

fig, ax = plt.subplots()
ax.plot(x, y, label='function')
ax.plot(x, zero, label='x-axis')
ax.set(title='2nd Root', xlabel='x', ylabel='y')
ax.legend(loc='best')


xmin = 3
xmax = 4
root = (xmax - xmin)/2 + xmin
while f(root) > 1e-10 or f(root) < -1e-10:
    root = (xmax - xmin)/2 + xmin
    if f(root) > 0:
        xmin = root
    elif f(root) < 0:
        xmax = root
    ax.axvline(root, color='black', linestyle='dashed')

print('2nd root: %s' %root)
```
