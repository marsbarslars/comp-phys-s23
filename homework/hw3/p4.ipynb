First, I will plot the quadratic equation.

```python
#Import necessary modules
import math
import numpy as np
import matplotlib.pyplot as plt

#Make array for x
#Make output array for y from x and equation
x = np.arange(-1100, 100, 1)
y = x**2 + 1000*x - 1

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(title='Quadratic Graph', xlabel='x', ylabel='y')
```

Next, I will find the smaller root with the quadratic formula.

```python
#Set coefficients to a, b, and c
a = 1
b = 1000
c = -1

#Use formula
root = (-b + math.sqrt(b**2 - 4*a*c))/(2*a)

print('Root by formula: %s' %root)
```

Next, I will use Newton's method the find the smaller root.

```python
#Define functions for y and derivative
def f(x):
    return x**2 + 1000*x - 1
def df(x):
    return 2*x + 1000

#Set up while loop to determine root with good precision
x = 0
while f(x) > 1e-10 or f(x) < -1e-10:
    x = x - f(x)/df(x)

print('Root by iteration: %s' %x)
```

Next, I will use the decimal module to redo the previous two parts.

```python
#Import module
from decimal import Decimal

#Set up and use the quadratic formula like above
a = Decimal(1)
b = Decimal(1000)
c = Decimal(-1)

root = (-b + Decimal.sqrt(b**Decimal(2) - Decimal(4)*a*c))/(Decimal(2)*a)

print('Root by formula: %s' %root)

#Define decimal functions
def decf(x):
    return x**2 + Decimal(1000)*x - Decimal(1)
def decdf(x):
    return Decimal(2)*x + Decimal(1000)

#Iterate using while loop like above
x = Decimal(0)
while decf(x) > Decimal(1e-10) or decf(x) < Decimal(-1e-10):
    x = x - decf(x)/decdf(x)

print('Root by iteration: %s' %x)
```

None of the 4 methods found the exact root of 0.001, but they were all very close. In both types of precision, finding the root by iterating was slower but more exact. With this small of a data set, however, there wasn't noticable lag. The iteration method was only more exact because I forced the output to be within certain, small bounds.

Although it is slight, there is a difference between both types of precision. Both ways of using the decimal module had a result that was pretty much the same as the iteration method using binary precision; the binary was just rounded up slightly. The difference was only really noticeable with the quadratic formula. Here, the decimal module was closer to the real root than the binary precision. This is because the decimal precision tries to be more lossless and is closer to the regular decimal numbers we people use.
