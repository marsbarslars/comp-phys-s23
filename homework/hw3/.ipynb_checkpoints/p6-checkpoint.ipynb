I'll start by plotting the function.

```python
#Import modules
import math
import numpy as np
import matplotlib.pyplot as plt

#Make x array and y array
x = np.arange(-1, 10, 0.01)
y = np.sin(math.pi + np.cos(x))

#Plot
zero = x * 0
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, zero)
ax.set(title='Graph of Function', xlabel='x', ylabel='y')
```

The first 3 roots are around $x = 1.5$, $x = 4.5$, and $x = 8$. Now, I'll find the first 3 roots using the midpoint method.

```python
#Define function
def f(x):
    return math.sin(math.pi + math.cos(x))

#This function finds the root of part with positive slope
def find_root_up(xmin, xmax):
    x = (xmax - xmin)/2 + xmin
    while f(x) > 1e-10 or f(x) < -1e-10:
        x = (xmax - xmin)/2 + xmin
        if f(x) < 0:
            xmin = x
        elif f(x) > 0:
            xmax = x
    return x
         
#This function finds the root of part with negative slope
def find_root_down(xmin, xmax):
    x = (xmax - xmin)/2 + xmin
    while f(x) > 1e-10 or f(x) < -1e-10:
        x = (xmax - xmin)/2 + xmin
        if f(x) > 0:
            xmin = x
        elif f(x) < 0:
            xmax = x
    return x

print('1st root: %s' %find_root_up(1, 2))
print('2nd root: %s' %find_root_down(4, 6))
print('3rd root: %s' %find_root_up(7, 9))
```

Going off the plot and these three roots, there seems to be a sequence of the roots being odd multiples of $\frac{\pi}{2}$.

```python
print('1st root: %s' %(math.pi/2))
print('2nd root: %s' %(3*math.pi/2))
print('3rd root: %s' %(5*math.pi/2))
```

These values match the roots as found above. Now, look back at the expression $$sin(\pi + cos(x)).$$ Plug in $\frac{n \pi}{2}$ for the x values, where n is any odd integer. $$sin(\pi + 0)$$ $$sin(\pi)$$ $$0$$ It works out!
