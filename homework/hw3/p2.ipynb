This program computes the sum relative to different $ \Delta x $.

```python
#Import necessary modules
import math
import numpy as np
import matplotlib.pyplot as plt

#Have epsilon from problem 1 ready
ep = 1.
while 1. + ep != 1.:
    inv = ep
    ep /= 2.
ep = inv

#Make/initiate array/list for plot
dx = np.arange(0.5, 1.001, 0.001)
zout = []

#Loop over dx array to find outputs
for x in dx:
    sum = 0
    for j in range(1000):
        sum += x
    z = 1 - (sum / (1000 * x))
    zout.append(z)

#Make output list into array and put into correct units
zerr = np.array(zout)
zerr = ep * zerr / 2

#Plot dx and zerr
fig, ax = plt.subplots(2, 1, figsize=(14, 10))
ax[0].plot(dx, zerr, 'o-')
ax[0].set(title='z Based on Rounding Error', xlabel=r'$\Delta x$', ylabel=r'$\frac{z}{\epsilon / 2}$')
ax[1].hist(zerr, bins=41)
ax[1].set(title='Histogram of z', xlabel=r'$\frac{z}{\epsilon / 2}$', ylabel='Frequency')
```
