This program computes the sum relative to different $ \Delta x $.

```python
#Import necessary modules
import math
import numpy as np
import matplotlib.pyplot as plt

#Have epsilon from problem 1 ready
ep = 1.
while 1. + ep != 1.:
    inv = ep
    ep /= 2.
ep = inv

#Make/initiate array/list for plot
dx = np.arange(0.5, 1.001, 0.001)
zout = []

#Loop over dx array to find outputs
for x in dx:
    sum = 0
    for j in range(1000):
        sum += x
    z = 1 - (sum / (1000 * x))
    zout.append(z)

#Make output list into array and put into correct units
zerr = np.array(zout)
zerr = ep * zerr / 2

#Plot dx and zerr
zero = dx * 0
fig, ax = plt.subplots(2, 1, figsize=(14, 10))
ax[0].plot(dx, zerr, 'o-')
ax[0].plot(dx, zero, color='black')
ax[0].scatter()
ax[0].set(title='z Based on Rounding Error', xlabel=r'$\Delta x$', ylabel=r'$\frac{z}{\epsilon / 2}$')
ax[1].hist(zerr, bins=41)
ax[1].set(title='Histogram of z', xlabel=r'$\frac{z}{\epsilon / 2}$', ylabel='Frequency')
```

There is no good correspondence between $\Delta x$ and the error (1st graph). However, there is a bimodal distribution of how often different values of the error occur. Much of the error is around the + or - $\frac{z}{\epsilon / 2} \leq error \leq \frac{z}{\epsilon}$ times the scale of the graph. There is less error in the area around 0. It makes sense that the distribution is symmetric because if there is error, there should be about the same chance of error in either the positive or negative direction. It's rare that the binomial and decimal representations of numbers are exact, so it makes sense that the magnitude of error is usually greater than 0.
