I will use the acceptance-rejection method to find the random numbers. First is finding the CDF $P(x) = R$. Then, I'll complete the square to find the positive inverse.
$$\newline$$
$$P(x) = \int_{0}^{x} \frac{1+4x'}{3} \,dx'$$
$$\newline$$
$$P(x) = \frac{x'}{3} + \frac{2x'^2}{3} |_{0}^{x}$$
$$\newline$$
$$R = \frac{2}{3} (x^2 + \frac{x}{2} + \frac{1}{16} - \frac{1}{16})$$
$$\newline$$
$$R = \frac{2}{3} (x + \frac{1}{4})^2 - \frac{1}{24}$$
$$\newline$$
$$x + \frac{1}{4} = \sqrt{\frac{3R}{2} + \frac{1}{16}}$$
$$\newline$$
$$x = \sqrt{\frac{3R}{2} + \frac{1}{16}} - \frac{1}{4}$$
$$\newline$$

On the histogram, $A$ should be $10^3$. The total area should be the total probability (1) times $10^4$. The pdf $f(x)$ is already normalized. On top of that, there are 10 bins, so that total area is scaled down by a factor of 10.

```python
#Import modules
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt

#Make functions
def f(x):
    return (1+4*x)/3

def pdf(R):
    return np.sqrt(3*R/2 + 1/16) - 1/4

#Make distribution and plot histogram
R = rand.uniform(0, 1, 10**4)
x = pdf(R)

fig, ax = plt.subplots(figsize=(10,10))
ax.hist(x, bins=10)

#Plot g(x) curve
y = np.linspace(0, 1, 1000)
ax.plot(y, f(y)*(10**3))
ax.set(xlabel='x', ylabel='g(x)')
plt.show()
```
