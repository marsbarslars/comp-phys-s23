First, I'll make a set of X values following the steps in the homework.

```python
#Import modules
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
import statistics as stat

#Build X using outside for loop and inside while loop to make sure u < 1
x = []
for j in range(10**6):
    i = 0
    u = 0
    while u < 1:
        u += rand.random(1)
        i += 1
    x.append(i)
```

Next, I'll simply make a histogram.

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(x, bins=range(min(x), max(x) + 1, 1))
ax.set(xlabel='X', ylabel='Frequency')
plt.show()
```

Find e and the error.

```python
est = sum(x)/len(x)

print('Estimate: ' + str(est))
print('Real Value: ' + str(np.exp(1)))
print('Variance: ' + str(stat.variance(x)))
print('Standard Deviation: ' + str(np.sqrt(stat.variance(x))))
```

Use the antithetic set to cut down the error.

```python
#Follow similar process for antithetic as I did for X
anti = []
for j in range(10**6):
    i = 0
    u = 0
    while u < 1:
        u += 1 - rand.random(1)
        i += 1
    anti.append(i)

x = np.array(x)
anti = np.array(anti)
est_list = (x + anti)/2
est = sum(est_list)/len(est_list)

print('New estimate: ' + str(est))
print('Variance: ' + str(stat.variance(est_list)))
print('Standard Deviation: ' + str(np.sqrt(stat.variance(est_list))))
```

It's closer to e this time around. The error got cut down by about half.
