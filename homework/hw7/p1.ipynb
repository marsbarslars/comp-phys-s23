This problem uses the fourth order Runge-Kutta method to find the output voltage of a low-pass filter.

```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt

#Make a function for Runge-Kutta for this case
def RK4(yn, dt, v_in, tau, f):
    k1 = dt*f(yn, v_in, tau)
    k2 = dt*f(yn + k1/2, v_in, tau)
    k3 = dt*f(yn + k2/2, v_in, tau)
    k4 = dt*f(yn + k3, v_in, tau)
    
    #Return y_{n+1} as new y_n
    yn = yn + (k1 + 2*k2 + 2*k3 + k4)/6
    return yn

#Create function for dV_out/dt
def f(v_out, v_in, tau):
    return (v_in - v_out)/tau

#Make "solve" function for problem
def solve_vout(tau):
    dt = 0.01
    t = np.arange(0, 100+dt/2, dt)
    v_out = np.empty(10**4 + 1)
    v_out[0] = 0
    
    v_in = np.empty(10**4 + 1)
    v_in[0] = 1
    
    #For loop to make arrays that I plot
    for i in range(1, 10**4 + 1):
        
        #Set v_in based on square wave
        if np.floor(t[i]) % 2 == 0:
            v_in[i] = 1
        else:
            v_in[i] = -1
        
        v_out[i] = RK4(v_out[i-1], dt, v_in[i], tau, f)
    return t, v_in, v_out

#Run function for tau = 0.1 microseconds
#v_in was just for testing
t, v_in, v_out = solve_vout(0.1)
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

ax[0].plot(t[:10**3], v_out[:10**3])
ax[0].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

ax[1].plot(t[10**4 - 10**3:], v_out[10**4 - 10**3:])
ax[1].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

plt.show()
```

When $\tau$ is very low, the cutoff frequency $\frac{1}{\tau}$ is very high, so most of the input voltage shows up as output as well.

```python
#Run function for tau = 1 microsecond
t, v_in, v_out = solve_vout(1)
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

ax[0].plot(t[:10**3], v_out[:10**3])
ax[0].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

ax[1].plot(t[10**4 - 10**3:], v_out[10**4 - 10**3:])
ax[1].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

plt.show()
```

When $\tau$ is at a middle value (the cutoff frequency is slightly above the frequency of the input voltage), the transient term shows as it quickly dies down. The peaks in the output voltage are significantly less than 1 because the output voltage doesn't have enough time to keep up with the input voltage.

```python
#Run function for tau = 4 microseconds
t, v_in, v_out = solve_vout(4)
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

ax[0].plot(t[:10**3], v_out[:10**3])
ax[0].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

ax[1].plot(t[10**4 - 10**3:], v_out[10**4 - 10**3:])
ax[1].set(xlabel=r't ($\mu$s)', ylabel=r'$V_{out}$ (V)')

plt.show()
```

When $\tau$ is high, the cutoff frequency is significantly lower than the frequency of the input voltage. The transient term sticks around for longer, and the output voltage never gets very high. Because of that, it seems to be like a triangle wave.
