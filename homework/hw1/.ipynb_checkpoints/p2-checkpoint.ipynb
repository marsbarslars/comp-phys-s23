This program graphs two decaying sine waves and their envelope.

```python
#Import all necessary modules
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#Create array where t ranges from 0 to 6pi
#Include 20 steps for each step of pi
#Create arrays for a, b, and c = sqrt(a^2+b^2)
t = np.linspace(0, 6*math.pi, 121)
a = np.exp(-t/8)*np.sin(t)
b = np.exp(-t/8)*np.cos(t)
c = (a**2 + b**2)**0.5

#Plot all three equations on the same graph
plt.plot(t,a, label=r'$a=e^{-t/8}sin(t)$')
plt.legend(loc='best')
plt.xlabel('t')

plt.plot(t,b, label=r'$b=e^{-t/8}cos(t)$')
plt.plot(t,c, label=r'$\sqrt{a^2+b^2}$')
```
