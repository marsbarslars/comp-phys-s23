This program shows a color map of a particle's probability distribution in a 2-D square well. The well is in the ranges 0<x<L and 0<y<L.

```python
#Import necessary modules
import math
import numpy as np
import matplotlib.colors as col
from matplotlib import cm
import matplotlib.pyplot as plt

#Make arrays of indep. variables
x = np.linspace(0, 1, 101)
y = np.linspace(0, 1, 101)

#Use array magic to simplify whole equation
x2 = np.sin(2*math.pi*x)
y2 = np.sin(5*math.pi*y)

#Make a 2-D array with psi values
#1st dimension is y-values
#2nd dimension is x-values
#Square result (all values are real) to get final array
psi = []
psi_sub = []
for j in range(len(y)):
    for i in range(len(x)):
        psi_sub.append(x2[i]*y2[j])
    psi.append(psi_sub)
    psi_sub = []
psi = np.array(psi)
psi2 = psi**2

#Make and save color map
plt.pcolormesh(psi2)
plt.savefig('color_map.pdf')
```
