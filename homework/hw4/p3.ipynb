First up is writing a function that calculates the integral with Simpson's method with 1000 points.

```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt

#Make integrand function
def f(m, x, th):
    return np.cos(m*th - x*np.sin(th))

#Make Bessel function with Simpson's method
def J(m, x):
    integral = 0
    N = 1000
    h = np.pi/N
    for i in range(N):
        integral += h/2 * (f(m, x, i*h)/3 \
        + f(m, x, (i+0.5)*h)*4/3 + f(m, x, (i+1)*h/3))
    return integral/np.pi

#Make x array and plot
x = np.linspace(0, 20, 1000)

fig, ax = plt.subplots(figsize = (10, 10))
ax.plot(x, J(0, x), label=r'$m=0$')
ax.plot(x, J(1, x), label=r'$m=1$')
ax.plot(x, J(2, x), label=r'$m=2$')

ax.set(xlabel='x', ylabel='J(m,x)')
ax.legend(loc='best')
plt.show()
```

Next up is making a density plot of the intensity of the diffraction pattern.

```python
#Make intensity function
def I(r, ld):
    k = 2*np.pi/ld
    return (J(1, k*r)/(k*r))**2

x = np.arange(-1e-6, 1e-6, 1e-8)
y = x

#Make square density plot
z = []
z_sub = []
for j in range(len(y)):
    for i in range(len(x)):
        if x[i] == 0 and y[j] == 0:
            z_sub.append(0.25)
        else:
            z_sub.append(I((x**2 + y**2)**0.5, 5e-7))
    z.append(z_sub)
    z_sub = []
z = np.array(z)
fig, ax = plt.subplots()
den = ax.contourf(x, y, z)
plt.colorbar(den)
plt.show()
```
