Consider
$$\newline$$
$$f(x) = \int_{0}^{\infty} e^{-(x+y)} (x+y)^N G(y|\mu, \sigma) \,dy$$
$$\newline$$
where
$$\newline$$
$$G(y|\mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(y-\mu)^2}{2\sigma^2}\right)$$
$$source: Wikipedia$$

A change of variable will make this a proper integral. Let $z = \frac{y}{1+y}$, so $y = \frac{z}{1-z}$ and $dy = \frac{1}{(1-z)^2} dz$.
$$\newline$$
$$f(x) = \int_{0}^{1} \frac{1}{(1-z)^2} e^{-\left(x+\frac{z}{1-z}\right)} \left(x+\frac{z}{1-z}\right)^N G\left(\frac{z}{1-z}|\mu, \sigma\right) \,dz$$

```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand

#Make x array
x = np.linspace(-3, 15, 100)

#Set constants
N = 5
mu = 3
sigma = 0.5

#Set up function
def part1(z):
    return 1/(1-z)**2

def part2(x, z):
    return np.exp(-x - z/(1-z))

def part3(x, N, z):
    return (x + z/(1-z))**N

def part4(x, mu, sigma, z):
    return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-0.5*((z/(1-z)-mu)/sigma)**2)

def f(x):
    output = []
    for i in range(100):
        z = rand.uniform(0, 1, 10**6)
        func = part1(z) * part2(x[i], z) * part3(x[i], 5, z) * part4(x[i], 3, 0.5, z)
        output.append(np.mean(func))
    return output

#Plot f(x)
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, f(x))
ax.set(xlabel='x', ylabel='f(x)')
plt.show()
```
